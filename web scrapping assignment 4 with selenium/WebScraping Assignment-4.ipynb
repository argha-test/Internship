```python
#webdriver
driver=webdriver.Chrome(r"D:/chromedriver/chromedriver.exe")
```

```python
#get the website
url="https://github.com/trending"
driver.get(url)
```

```python
links=[]
try:
    for l in driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a'):
        links.append(l.get_attribute('href'))
except NoSuchElementException:
        links.append("-")
links
```

```python
Repository_title=[]
Repository_description=[]
Contributors_count=[]
Language_used=[]
a="--"

for i in links:
    driver.get(i)
    
    #scraping the titles
    try:
        t = driver.find_element_by_xpath('//*[@id="js-repo-pjax-container"]/div[1]/div[1]/div/h1').text
        Repository_title.append(t.replace("\n",""))
    except NoSuchElementException:
        Repository_title.append(a)

    #scraping the description
    try:
        d = titles = driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/p').text
        Repository_description.append(d) 
    except NoSuchElementException:
        Repository_description.append(a)    
                        
    #scraping the company_name 
    try:
        #c = driver.find_element_by_xpath('//div[@class="flex-shrink-0 col-12 col-md-3"]/div/div[3]/div/h2/a/span').text
        c= driver.find_element_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div[2]/div/h2/a/span').text
        Contributors_count.append(c) 
    except NoSuchElementException:
        Contributors_count.append(a)

        
    #scraping the Language
    try:
        l1 = driver.find_element_by_xpath('//a[@class="d-inline-flex flex-items-center flex-nowrap Link--secondary no-underline text-small mr-3"]').text
        Language_used.append(l1.replace("\n"," - ")) 
    except NoSuchElementException:
        Language_used.append(a)
```

```python
print(len(Repository_title),len(Repository_description),len(Contributors_count),len(Language_used))
```

```python
Github=pd.DataFrame({"Repository title":Repository_title, "Repository description":Repository_description,
                    "Contributors count":Contributors_count,"Language used":Language_used})
Github
```
