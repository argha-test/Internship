Data Collection Phase: In this section, you need to scrape images from e-commerce portal, Amazon.com. The clothing categories used for scraping will be:

- Sarees (women)
- Trousers (men)
- Jeans (men)

```python
# Importing Libraries
import selenium
import pandas as pd
import time

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

#Importing warings
import warnings
warnings.filterwarnings('ignore')
```

```python
#webdriver
driver=webdriver.Chrome(r"D:\chromedriver\chromedriver.exe")
time.sleep(3)
```

- Sarees (women)

```python
#opening the homepage of amazon.in
driver.get('https://www.amazon.in/s?rh=n%3A1968256031&fs=true&ref=lp_1968256031_sar')
```

```python
urls=[]
max_page=7 #maximum number of pages to extract
max_p_d=2 #page number digits
for i in range (2, max_page +1):
    url=f"https://www.amazon.in/s?i=apparel&rh=n%3A1968256031&fs=true&page={i}&qid=1632122277&ref=sr_pg_{i}"
    urls.append(url)
    
urls
```

```python
links1=[]
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
time.sleep(2)
try:
    for l in driver.find_elements_by_xpath('//*[@class="a-section aok-relative s-image-tall-aspect"]/img'):
        if l.get_attribute('src') and 'https' in l.get_attribute('src'):
            links1.append(l.get_attribute('src'))
except NoSuchElementException:
    links1.append("-")
for i in urls:
    driver.get(i)
    try:
        for l in driver.find_elements_by_xpath('//*[@class="a-section aok-relative s-image-tall-aspect"]/img'):
            if l.get_attribute('src') and 'https' in l.get_attribute('src'):
                links1.append(l.get_attribute('src'))
    except NoSuchElementException:
        links1.append("-")
```

```python
len(links1)
```

```python
import os
import io
import requests
from PIL import Image

for i, url in enumerate(links1):
    file_name = f"S{i}.jpg"    
    
    if i <=280:
        os.chdir(r'D:\Clothing\train\Saree')
        baseDir=os.getcwd()
        try:
            image_content = requests.get(url).content

        except Exception:
            print(f"ERROR - unable to download {url} \n")

        try:
            image_file = io.BytesIO(image_content)
            image = Image.open(image_file).convert('RGB')

            loc = os.path.join(baseDir, file_name)

            with open(loc, 'wb') as f:
                image.save(f, "JPEG", quality=85)
            print(f"SAVED - {url} - AT: {loc}")
        except Exception:
            print(f"ERROR - unable to download {url} \n")
    elif i <=350:
        os.chdir(r'D:\Clothing\validate\Saree')
        baseDir=os.getcwd()
        try:
            image_content = requests.get(url).content

        except Exception:
            print(f"ERROR - unable to download {url} \n")

        try:
            image_file = io.BytesIO(image_content)
            image = Image.open(image_file).convert('RGB')

            loc = os.path.join(baseDir, file_name)

            with open(loc, 'wb') as f:
                image.save(f, "JPEG", quality=85)
            print(f"SAVED - {url} - AT: {loc}")
        except Exception:
            print(f"ERROR - unable to download {url} \n")
    else:
        os.chdir(r'D:\Clothing\test')
        baseDir=os.getcwd()
        try:
            image_content = requests.get(url).content

        except Exception:
            print(f"ERROR - unable to download {url} \n")

        try:
            image_file = io.BytesIO(image_content)
            image = Image.open(image_file).convert('RGB')

            loc = os.path.join(baseDir, file_name)

            with open(loc, 'wb') as f:
                image.save(f, "JPEG", quality=85)
            print(f"SAVED - {url} - AT: {loc}")
        except Exception:
            print(f"ERROR - unable to download {url} \n")
```
